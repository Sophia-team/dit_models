```python
import pandas as pd
```

```python
DATA_FOLDER = "./data/"
DENSITY03 = "02_CDensity_1_201903.csv.gz"
DENSITY04 = "02_CDensity_1_201904.csv.gz"
```

# Density03

```python
density03 = pd.read_csv(DATA_FOLDER + DENSITY03, sep=";", compression="gzip")
```

```python
density03.head()
```

```python
target_df = pd.read_csv(DATA_FOLDER + "address_cell_id.csv")
```

```python
density03['ts'][0][3:5]
```

```python
target_df
```

```python
density03['month'] = 3
density03['day'] = density03.ts.transform(lambda x: int(x[3:5]))
```

```python
density03
```

```python
agg_density03 = density03.groupby(["day", "zid"])[["customers_cnt_total",
                                   "customers_cnt_long",
                                   "customers_cnt_work",
                                   "customers_cnt_loc"]].mean()
```

```python
agg_density03 = agg_density03.reset_index()
```

```python
agg_density03['month'] = 3
```

```python
agg_density03.to_parquet(DATA_FOLDER + "aggregated_density03.parquet")
```
